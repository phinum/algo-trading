Volatility Indicator

Bollinger Bands comprises of 2 lines plotted n (typically 2) standard deviation away from an m (typically 20) period simple moving average line. The bands widen during increased volatility period and shrink during decreased volatility period.

```python
import pandas_datareader.data as pdr
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tick = "SBIN.NS"
df = pdr.get_data_yahoo(tick,start="2015-05-01", end="2020-05-1")
df
```

```python
df['Moving Average'] = df['Adj Close'].rolling(20).mean()
df
```

```python
df['BB-up'] = df['Moving Average'] + 2*df['Adj Close'].rolling(20).std()
df
```

```python
df['BB-down'] = df['Moving Average'] - 2*df['Adj Close'].rolling(20).std()
df
```

```python
df['BB-width'] = df['BB-up'] - df['BB-down']
df
```

```python
plt.figure(figsize = (10,6))
plt.plot(df.iloc[-200:,[-4,-3,-2]])
plt.title('Bollinger Bands')
plt.legend(('Moving Average','BB-up','BB-down'),loc='best')
plt.show()
```
